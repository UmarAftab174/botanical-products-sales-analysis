```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Preprocessig Data

#### Import seperate excel sheets as three different dataframes

```python
sales_df = pd.read_excel('Plant_DTS.xlsx', sheet_name='Plant_FACT')
delivery_df = pd.read_excel('Plant_DTS.xlsx', sheet_name='Accounts')
product_df = pd.read_excel('Plant_DTS.xlsx', sheet_name='Plant_Hierarchy')
```

#### Merging 3 sheets into one dataframe on basis of Product ID and Account ID

```python
product_df.rename(columns={'Product_Name_id' : 'Product_id'}, inplace=True)

plants_df = pd.merge(
    (pd.merge(sales_df, delivery_df, how='inner',on='Account_id')),
    product_df, 
    how='inner', 
    on='Product_id'
)
```

#### Modifying columns to make them suitable for EDA

```python
plants_df.columns = plants_df.columns.str.lower()
plants_df.columns
```

```python
plants_df.rename(columns={'latitude2' : 'latitude','country2' : 'country','cogs_usd' : 'cost_of_goods_usd'}, inplace=True)
```

# Descrptive Analysis

```python
plants_df.describe()
```

#### Working on feasibility and profitability of product family

```python
print(plants_df['product_family'].value_counts())
```

```python
prod_family_group = plants_df.groupby(['product_family'])
prod_family_metrics = prod_family_group[['cost_of_goods_usd','price_usd','sales_usd','quantity']].sum()
```

```python
prod_family_finance = prod_family_metrics.copy()
prod_family_finance['profit'] = prod_family_finance['sales_usd'] - prod_family_finance['cost_of_goods_usd']
prod_family_finance['profit_margin'] = (prod_family_finance['profit'] / prod_family_metrics['sales_usd'] * 100)
```

```python
prod_family_finance = prod_family_finance.round(4)
```

```python
prod_family_finance.reset_index(inplace=True)
prod_family_finance.to_csv("product_family_performance.csv", index=False)
```

#### Analysis of country and account's finances

```python
print(plants_df['country'].value_counts())
print(plants_df['account'].value_counts())
```

```python
plants_df['year'] = plants_df['date_time'].dt.year
plants_df['quarter'] = plants_df['date_time'].dt.quarter

account_country_group = plants_df.groupby(['country', 'year', 'quarter', 'account'])[['cost_of_goods_usd','price_usd','sales_usd','quantity']].sum()
account_country_group
```

```python
account_country_group['avg_price_per_item'] = account_country_group['sales_usd'] / account_country_group['quantity']
account_country_group['markup_ratio'] = account_country_group['price_usd'] / account_country_group['cost_of_goods_usd']
```

```python
account_country_group = account_country_group.round(4)
```

```python
account_country_group.reset_index(inplace=True)
account_country_group.to_csv('account_country_performance.csv', index=False)
```
